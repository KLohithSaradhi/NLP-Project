```python
df = pd.read_csv("./ProcessedDanceScripts.csv")
```

```python
import gensim
```

```python
sentences = list(map(lambda x : eval(x), list(df["lems"])))
```

```python
mod = gensim.models.FastText(sentences, window=5, vector_size = 5, workers=4, sg=1)
```

```python
mod.wv.most_similar("dance", topn=100)
```
